```python
# Cell 1
import pandas as pd, sys; sys.path.append("..")
from src.cohort_rfm import build_cohort_retention, compute_rfm, plot_cohort_heatmap, plot_rfm_segments

df = pd.read_csv("../data/raw/sessions.csv")
```

```python
# Cell 2 — Cohort heatmap
retention = build_cohort_retention(df)
display(retention.round(1))
plot_cohort_heatmap(retention, save_path="../outputs/figures/cohort_heatmap.png")
```

```python
# Cell 3 — RFM
rfm = compute_rfm(df)
display(rfm.groupby("segment")[["recency","frequency","monetary"]].mean().round(2))
plot_rfm_segments(rfm, save_path="../outputs/figures/rfm_segments.png")
```

```python

# Cell 4 — Save
rfm.to_csv("../data/processed/rfm_scores.csv", index=False)
retention.to_csv("../data/processed/cohorts.csv")
```
